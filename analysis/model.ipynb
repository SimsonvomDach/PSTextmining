# This is the Script for the Word2Vec Model with the gensim library

```python
import pandas as pd
import nltk
```

```python
df = pd.read_csv("NER.csv", index_col=0)
df
```

```python
def preprocess(row):
    # Text to be preprocessed
    text = row['Dialogue']
    
    # Tokenize the text
    tokens = nltk.word_tokenize(text)
    
    # POS tag the tokens
    pos_tags = nltk.pos_tag(tokens)
    
    # NER tag the tokens
    ner_tags = nltk.ne_chunk(pos_tags)
    
    # Preprocess the tokens, POS tags, and NER tags to create a list of input sequences
    input_sequences = []
    input_sequence = []
    for token, pos_tag, ner_tag in zip(tokens, pos_tags, ner_tags):
        input_sequence.append(token)
        if token == '.':
            input_sequences.append(input_sequence)
            input_sequence = []
    
    # Return the input sequences
    return input_sequences

# Preprocess the text data in the 'text' column of the data frame
df['input_sequences'] = df.apply(preprocess, axis=1)
```
